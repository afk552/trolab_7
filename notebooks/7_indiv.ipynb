Выделить прямоугольником на изображении всех людей красным цветом и объекты - зеленым. 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
def imgshow(image, conversion=cv2.COLOR_BGR2RGB):
    image = cv2.cvtColor(image, conversion)
    # Show the image
    plt.imshow(image)
    # remove the axis / ticks for a clean looking image
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
image = cv2.imread('pictures/people.jpg')
print(image.shape)
```

```python
image = cv2.rectangle(image,(406,194),(438,236),(0,0,255),3)
image = cv2.rectangle(image,(96,207),(135,243),(0,0,255),3)
image = cv2.rectangle(image,(244,181),(306,240),(0,0,255),3)
image = cv2.rectangle(image,(517,169),(555,224),(0,0,255),3)
image = cv2.rectangle(image,(153,193),(205,255),(0,0,255),3)
image = cv2.rectangle(image,(675,270),(699,313),(0,255,0),3)
image = cv2.rectangle(image,(58,222),(86,253),(0,255,0),3)
imgshow(image)
```
